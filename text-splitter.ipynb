```python
from langchain_text_splitters import CharacterTextSplitter
```

```python
# This is a long document we can split up.
with open("dictionary.txt") as f:
    disease_synonyms = f.read()
```

```python
text_splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=63000,
    chunk_overlap=0,
    length_function=len,
    is_separator_regex=False,
)
```

```python
from pathlib import Path
import shutil
outdir = Path('disease_syn_documents')
if outdir.exists():
    shutil.rmtree(outdir)
    outdir.mkdir()
if not outdir.exists():
    outdir.mkdir()
texts = text_splitter.create_documents([disease_synonyms])
for idx, doc in enumerate(texts):
    with open(outdir / f"doc_{idx}.txt", "w") as f:
        f.write(doc.page_content)
        print(len(doc.page_content), "written")
```
